```python
import os

import pandas as pd
import numpy as np


from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
import plotly.express as px
import plotly.graph_objs as go
import matplotlib.pyplot as plt

init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('Desktop/owid_covid_data.csv')
```

```python
df.tail()
```

```python
df['death_rate'] = (df['new_deaths_smoothed_per_million'] / df['new_cases_smoothed_per_million']).replace(np.inf,np.nan)
df['population_coverage'] = df['total_tests'] / df['population']
df['datetime'] = pd.to_datetime(df['date'])
```

```python
df.describe().T
```

```python
df.info()
```

```python
problem_idx = df[(df['new_cases']<0)|(df['new_deaths']<0)|(df['new_cases_smoothed']<0)|(df['new_deaths_smoothed']<0)|(df['new_cases_per_million']<0)|(df['new_deaths_per_million']<0)
                 |(df['new_deaths_smoothed_per_million']<0)|(df['new_tests']<0)|(df['new_tests_per_thousand']<0)|(df['location'].isin(['World']))].index
```

```python
df = df[~df.index.isin(problem_idx)]
df = df[~df['date'].isin(['2020-09-18', '2020-09-19'])].reset_index(drop=True)
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
def impute_median(total_cases):
    return total_cases.fillna(total_cases.median())
df.total_cases = df['total_cases'].transform(impute_median)

def impute_median(death_rate):
    return death_rate.fillna(death_rate.median())
df.death_rate = df['death_rate'].transform(impute_median)

def impute_median(population_coverage):
    return population_coverage.fillna(population_coverage.median())
df.population_coverage = df['population_coverage'].transform(impute_median)
```

```python
df.isnull().sum()
```

```python
df.columns
df2=df.copy()
df2=df2.drop(df2.columns[0],axis=1) #Dropping the unnamed column
df2=df2.drop(columns=['reproduction_rate','icu_patients','icu_patients_per_million','handwashing_facilities',
                      'human_development_index','life_expectancy','extreme_poverty'])
```

```python
trace1 = go.Scatter(
    x=df.groupby(['date'])['date'].apply(lambda x: np.unique(x)[0]),
    y=df.groupby(['date'])['new_tests_smoothed'].sum().astype(int),
        xaxis='x2',
    yaxis='y2',
    name = "new tests smoothed"
)
trace2 = go.Scatter(
    x=df2.groupby(['date'])['date'].apply(lambda x: np.unique(x)[0]),
    y=df2.groupby(['date'])['new_deaths_smoothed'].sum().astype(int),
    name = "new deaths smoothed"
)
trace3 = go.Scatter(
    x=df2.groupby(['date'])['date'].apply(lambda x: np.unique(x)[0]),
    y=(df2.groupby(['date'])['positive_rate'].mean() * 100).round(3),
    xaxis='x3',
    yaxis='y3',
    name = "test positive rate %"
    )
trace4 = go.Scatter(
    x=df2.groupby(['date'])['date'].apply(lambda x: np.unique(x)[0]),
    y=df2.groupby(['date'])['new_cases_smoothed'].sum().astype(int),
    xaxis='x4',
    yaxis='y4',
    name = "new cases smoothed"
)

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'New tests, deaths, cases and test positive rate'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
fig = px.scatter(df2_grouped, 
                 x="new_deaths", y="gdp_per_capita", size="new_cases", color="continent",
                 hover_name="location", log_x=True, size_max=60)
fig.show()
```
